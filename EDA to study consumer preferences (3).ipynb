```python
cat=[1,4] #array containing index of categorical features
```

```python
#function to remove all the outliers

def remove_outlier(data,cat):
    c=len(data.columns)
    threshold=3
    for x in range(c):
        if x not in cat:
            y=data.iloc[:,x]
            mean=np.mean(y)
            std=np.std(y)
            z_score=(y-mean)/std
            data=data[np.abs(z_score)<threshold]
    return data 
```

```python
data=remove_outlier(data,cat)
```

### Customer Segmentation using Kmeans and Elbow Method

```python
data2=data.copy()
```

```python
data2.shape
```

```python
data2.columns
```

```python
#dropping the unnecessary features
data2.drop(['Year_Birth','Marital_Status','Kidhome','Dt_Customer'],axis=1,inplace=True)
```

```python
from sklearn.cluster import KMeans
```

```python
#using elbow method to find the optimum no. of clusters
wcss=[]
for i in range(1,20):
    kmean=KMeans(n_clusters=i, init='k-means++', random_state=30)
    kmean.fit(data2)
    wcss.append(kmean.inertia_)
```

```python
plt.plot(range(1,20),wcss)
```

The optimum no. of clusters are-3

```python
kmean=KMeans(n_clusters=3, init='k-means++', random_state=30)
kmean.fit(data2)
```

```python
data2['cluster no.']=kmean.fit_predict(data2)
```

```python
data2
```

```python
data2=data2.groupby(['cluster no.']).mean().round()
data2=data2.reset_index()
```

```python
data2 # showing data according to mean value for all features for the 3 clusters
```
